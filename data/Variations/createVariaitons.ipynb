```python
import os
import shutil
```

```python
real_datadir = "/coc/flash5/mpatel377/repos/SLaTe-PRO/data/RoboCraft/processed_seqLM_coarse/test"
real_data = [os.path.join(real_datadir, d) for d in os.listdir(real_datadir)]

gpt_datadir = "/coc/flash5/mpatel377/repos/SLaTe-PRO/data/GPTdataGeneration/processed_seqLM_coarse/test"
gpt_data = [os.path.join(gpt_datadir, d) for d in os.listdir(gpt_datadir)]

print(len(real_data), len(gpt_data))
```

## GPT only cross validation

```python
num_splits = 5
num_data_in_split = len(gpt_data) // num_splits

print(f"Creating {num_splits} splits of {num_data_in_split} test points each")

for split in range(num_splits):
    split_dir = f"gpt_only_split_{split}/processed_seqLM_coarse"
    test_filepaths = gpt_data[split*num_data_in_split:(split+1)*num_data_in_split]
    train_filepaths = gpt_data[:split*num_data_in_split] + gpt_data[(split+1)*num_data_in_split:]
    os.makedirs(split_dir, exist_ok=True)
    os.makedirs(os.path.join(split_dir, 'train'), exist_ok=True)
    os.makedirs(os.path.join(split_dir, 'test'), exist_ok=True)
    shutil.copy(os.path.join('/coc/flash5/mpatel377/repos/SLaTe-PRO/data/GPTdataGeneration/processed_seqLM_coarse/common_data.json'), os.path.join(split_dir, 'common_data.json'))
    shutil.copy(os.path.join('/coc/flash5/mpatel377/repos/SLaTe-PRO/data/GPTdataGeneration/processed_seqLM_coarse/common_edge_data.pt'), os.path.join(split_dir, 'common_edge_data.pt'))
    for filepath in test_filepaths:
        shutil.copy(filepath, os.path.join(split_dir, 'test', os.path.basename(filepath)))
    for filepath in train_filepaths:
        shutil.copy(filepath, os.path.join(split_dir, 'train', os.path.basename(filepath)))
```
